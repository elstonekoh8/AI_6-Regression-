```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
# Using pandas to view import and view data
data = np.loadtxt("ex1data1.txt", delimiter=',')
x = np.c_[data[:,0], np.ones(data.shape[0])] 
y = np.c_[data[:,1]]
```

```python
# Lets plot the data to see how they correlate.
plt.scatter(x[:,0], y, s=32, marker='o')
plt.xlabel("Population of city in 10,000")
plt.ylabel("Profit in $10,000")
```

```python
# Linear Model Function Using SciKlearn
def linear_model(x,y):
    from sklearn.linear_model import LinearRegression
    reg = LinearRegression().fit(x, y)
    theta=np.c_[reg.coef_[:,0],reg.intercept_[:]]
    return theta

```

```python
# Ploting Regression Line from given Sample Data
theta = linear_model(x, y)
print('Our learned value of theta: ',theta.ravel())
x_plot = np.arange(5,23)
y_plot = theta[:,0]*x_plot + theta[:,1]
plt.scatter(x[:,0], y, s=30, c='r', marker='o', linewidths=1)
plt.plot(x_plot,y_plot, label='Linear regression (Scikit Learn)')
```

```python
# Prediction of Profit Given Population Size

input_ = float(input("Input the population size/10000: "))
profit = theta[:,0]*input_ + theta[:,1]
print("For a population of ", input_, ", the profit is $", profit)
```
